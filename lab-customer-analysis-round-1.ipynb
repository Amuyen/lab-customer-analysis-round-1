```python
import pandas as pd
import numpy as np
```

Read the three files into python as dataframes

```python

org_file1= pd.read_csv('files_for_lab/csv_files/file1.csv')
org_file2= pd.read_csv('files_for_lab/csv_files/file2.csv')
org_file3= pd.read_csv('files_for_lab/csv_files/file3.csv')

file1 = org_file1.copy()
file2 = org_file2.copy()
file3 = org_file3.copy()

df_list = [file1,file2,file3]
```


Show the DataFrame's shape.

```python
for i in range(len(df_list)):
    print('shape of file'+str(i+1)+' is: ',df_list[i].shape)
```

Standardize header names.

```python

def transf(x):
        x = x.lower().lstrip().rstrip().replace(" ","_").replace("state","st")
        return x              

def column_ch(y):
        cols_dic={}
        for c in range(y.shape[1]):
               cols_dic[y.columns[c]]=transf(y.columns[c])
        return cols_dic

for df in df_list:
        df.rename(columns=column_ch(df),inplace=True)
        display(df.head(2))
```

Correct the values in the column 'customer lifetime value'. They are given as a percent, so divide them by 100 and change dtype to numerical type. (only 2 of the files need this step.)

```python
for df in df_list:
    if df['customer_lifetime_value'].dtype==object:
        df['customer_lifetime_value']=df['customer_lifetime_value'].str.replace("%","")
        df['customer_lifetime_value']=df['customer_lifetime_value'].astype(float)/100
```

Concatenate the three dataframes

```python
dfconc=pd.concat(df_list)
dfconc
```
